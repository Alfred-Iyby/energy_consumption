```python
import pandas as pd
import numpy, os
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import seaborn as sns,matplotlib.pyplot as plt
```

Phase 1 

```python
parent_directory = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
phase_1 =  pd.read_csv(parent_directory + '/data/phase_1_data.csv')
phase_1.head()
```

```python
X = phase_1[['Hour', 'Demand_kWh', 'Temperature', 'Temperature24', 'Temperature48', 'Temperature72']]

X = add_constant(X)

vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("Phase 1\n",vif_data)
```

Phase 1 with survey

```python
parent_directory = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
phase_1_survey =  pd.read_csv(parent_directory + '/data/phase_1_data_with_survey.csv')
phase_1_survey.head()
```

```python
phase_1_survey['home_size'] = phase_1_survey['home_size'].astype('category').cat.codes
phase_1_survey['electric_car'] = phase_1_survey['electric_car'].astype('category').cat.codes
phase_1_survey['electrically_heated'] = phase_1_survey['electrically_heated'].astype('int')
X = phase_1_survey[['Hour', 'Demand_kWh', 'Temperature', 'home_size', 'electric_car', 'no_of_people','electrically_heated']]

X = add_constant(X)

vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("Phase 1 Survey\n",vif_data)
```
