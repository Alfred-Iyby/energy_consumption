```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
parent_directory = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
phase_1 =  pd.read_csv(parent_directory + '/data/phase_1_data.csv')
phase_1.head()
```

```python
survey_phase_1 = pd.read_csv(parent_directory + '/data/survey1_answers.csv')
survey_phase_1.head()
```

```python
survey_phase_2 = pd.read_csv(parent_directory + '/data/survey2_answers.csv')
survey_phase_2.head()
```

```python
phase_1_ID = set(phase_1["ID"])
survey_phase_1_ID = set(survey_phase_1["ID"])

unique_ids_1 = phase_1_ID.intersection(survey_phase_1_ID)
unique_ids_1
```

```python
questions_selected = survey_phase_1[['ID','q5','q16_bil_b','q9N1','q9N2','q9N3','q9N4','q9N5','q9N6','q9N7','q9N8','q19']]
questions_selected = questions_selected.dropna()
questions_selected['q19'] = questions_selected['q19'].str.extract('(\d+)').astype(int)
questions_selected
```

```python
def is_electrically_heated(row):
    electric_sources = ['Main heat source', 'Additional heat source']
    if (row['q9N1'] in electric_sources) or (row['q9N2'] in electric_sources) or (row['q9N3'] in electric_sources):
        return True
    else:
        return False

questions_selected['electrically_heated'] = questions_selected.apply(is_electrically_heated, axis=1)
questions_selected=questions_selected.drop(columns=['q9N1','q9N2','q9N3','q9N4','q9N5','q9N6','q9N7','q9N8'])
questions_selected.rename(columns={'q5': 'home_size'}, inplace=True)
questions_selected.rename(columns={'q16_bil_b': 'electric_car'}, inplace=True)
questions_selected.rename(columns={'q19': 'no_of_people'}, inplace=True)
```

```python
phase_1_survey_merged = pd.merge(phase_1, questions_selected, on='ID', how='inner')
phase_1_survey_merged = phase_1_survey_merged.drop(columns=["Temperature24","Temperature48","Temperature72"])
phase_1_survey_merged.to_csv(parent_directory+"/data/phase_1_data_with_survey.csv",index=False)
phase_1_survey_merged
```

```python
phase_2 =  pd.read_csv(parent_directory + '/data/phase_2_data.csv')
phase_2.head()
```

```python
survey_phase_2 = pd.read_csv(parent_directory + '/data/survey3_answers.csv')
survey_phase_2.head()
```

```python
phase_2_ID = set(phase_2["ID"])
survey_phase_2_ID = set(survey_phase_2["ID"])

unique_ids_2 = phase_2_ID.intersection(survey_phase_2_ID)
unique_ids_2
```

```python
questions_selected = survey_phase_2[['ID','Q2','Q8','QI6','Q6_1','Q6_2','Q6_3','Q6_4','Q6_5','Q6_6','Q6_7','Q6_8','Q6_9','Q6_10']]
questions_selected = questions_selected.dropna()
questions_selected = questions_selected[questions_selected["Q2"]!="Don't know"]
questions_selected
```

```python
def is_electric_heated(row):
    electric_heating_keywords = [
    'Electric heaters', 
    'Electric underfloor heating', 
    'Air-to-air or air-to-water heat pump', 
    'Waterborne heat that is heated with electricity in your own home'
]
    return True if any(keyword in str(value) for keyword in electric_heating_keywords for value in row) else False

questions_selected['electrically_heated'] = questions_selected.apply(is_electric_heated, axis=1)
questions_selected = questions_selected.drop(columns=['Q6_1','Q6_2','Q6_3','Q6_4','Q6_5','Q6_6','Q6_7','Q6_8','Q6_9','Q6_10'])
questions_selected.rename(columns={'Q2': 'home_size'}, inplace=True)
questions_selected.rename(columns={'Q8': 'electric_car'}, inplace=True)
questions_selected.rename(columns={'QI6': 'no_of_people'}, inplace=True)

questions_selected
```
